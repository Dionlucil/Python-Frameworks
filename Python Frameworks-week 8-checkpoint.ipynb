## Part 2: Data Cleaning and Preparation

```python
# Drop columns with too many missing values (example threshold: > 80%)
missing_percentage = data.isnull().mean() * 100
to_drop = missing_percentage[missing_percentage > 80].index
data_clean = data.drop(columns=to_drop)
data_clean
```

```python
# Fill or drop missing values in important columns
data_clean1 = data_clean.dropna(subset=["title", "publish_time"])
data_clean1
```

```python
# Convert publish_time to datetime
data_clean1["publish_time"] = pd.to_datetime(data_clean1["publish_time"], errors="coerce")

# Extract year using .dt accessor
data_clean1["year"] = data_clean1["publish_time"].dt.year

# Now create data_clean2 as a copy
data_clean2 = data_clean1.copy()
data_clean2
```

```python
# Create abstract word count column on the cleaned dataset

data_clean2["abstract_word_count"] = data_clean2["abstract"].fillna("").apply(lambda x: len(x.split()))

# Preview
data_clean2[["abstract", "abstract_word_count"]].head()
```

## Part 3: Data Analysis and Visualization

```python
# Count papers by publication year

year_counts = data_clean2["year"].value_counts().sort_index()

plt.figure(figsize=(8,5))
plt.bar(year_counts.index, year_counts.values, color="skyblue")
plt.title("Publications by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
# top journals publishing COVID-19 research

top_journals = data_clean2["journal"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_journals.plot(kind="bar", color="orange")
plt.title("Top 10 Journals")
plt.xlabel("Journal")
plt.ylabel("Number of Publications")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Word frequency in titles (Word Cloud) 

titles = " ".join(data_clean2["title"].dropna().tolist())
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(titles)

plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Paper Titles")
plt.show()
```

```python
#  Distribution of papers by source 

if "source_x" in data_clean2.columns:  # safety check
    source_counts = data_clean2["source_x"].value_counts().head(10)

    plt.figure(figsize=(10,5))
    source_counts.plot(kind="bar", color="green")
    plt.title("Paper Counts by Source")
    plt.xlabel("Source")
    plt.ylabel("Number of Papers")
    plt.xticks(rotation=45, ha="right")
    plt.show()
```
